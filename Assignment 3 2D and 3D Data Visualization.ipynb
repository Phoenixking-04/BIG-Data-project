```python
import numpy as np
import matplotlib.pyplot as plt
#Let's establish the range for x by using intervals of π/4.
x_val = np.arange(-2 * np.pi, 2 * np.pi, np.pi / 4)
# Calculate the values of sine and cosine for the angle ( x ) and the angle (x + π/2)
sig_val = np.sin(x_val)
cos_val = np.cos(x_val + np.pi / 2)
# Create a line plot that includes both functions.
plt.figure(figsize=(12, 6))
plt.plot(x_val, sig_val, color='red', label='sin(x)')
plt.plot(x_val, cos_val, color='green', label='cos(x + π/2)')
plt.title('Line plot for both sin(x) and cos(x + π/2)')
plt.xlabel('X - Values')
plt.ylabel('Y - Values')
plt.legend()
plt.grid(True)
plt.show()
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))
ax1.plot(x_val, sig_val, color='red')
ax1.set_title('sin(x)')
ax1.grid(True)
ax2.plot(x_val, cos_val, color='green')
ax2.set_title('cos(x + π/2)')
ax2.grid(True)
plt.tight_layout()
plt.show()
```
