```python
import pandas as pnd
import numpy as np

DG_Data = pnd.read_csv('NCI_SEER_CRC.csv')
print(DG_Data)
```

```python
Data = DG_Data.rename(columns ={"Year of diagnosis":"YDD","Race recode (W, B, AI, API)":"Race",
                                "Origin recode NHIA (Hispanic, Non-Hisp)":"Ethnicity",
                                "Age recode with single ages and 100+":"Age","Grade (thru 2017)":"Grade",
                                "Marital status at diagnosis":"Marital Status"})

print(Data)
```

```python
columns= ['Age recode with <1 year olds','Primary Site - labeled','Total number of in situ/malignant tumors for patient',
               'COD to site recode','SEER registry (with CA and GA as whole states)','Behavior recode for analysis']
Data = Data.drop(columns=columns, axis = 1)

print(Data)
```

```python
Data['Age'] = Data['Age'].str.replace(' years', '')
print(Data)
```

```python
Data['Ethnicity'] = Data['Ethnicity'].replace({'Spanish-Hispanic-Latino' : 1, 'Non-Spanish-Hispanic-Latino' : 0})

print(Data)
```

```python
Data['Grade'] = Data['Grade'].replace({'Grade I' : 1,'Moderately differentiated; Grade II' : 2,'Poorly differentiated; Grade III' : 3,'Grade IV' : 4,'Unknown' : 99})
print(Data)
```

```python
Data['Sex'] = Data['Sex'].replace({'Female': 1, 'Male': 0})
print(Data)
```

```python
Data.dropna()
Data.fillna(0)
print(Data)
```
