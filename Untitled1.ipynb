```python
import pandas as pd

# Read the dataset
df = pd.read_csv('SEER.csv')

# Specify the correct columns to be included
columns_to_include = [
    'Age with single ages and 100+',
    'Year of diagnosis',
    'Sex',
    'Race',
    'Ethnicity',
    'Marital status at diagnosis',
    'Primary Site',
    'Summary Stage (2004+)',
    'Survival recode',  # Ensure to include the target variable
    'Appalachia'
]

# Drop all other features
df = df[columns_to_include]

# Divide the dataset into two parts
data_A = df[df['Appalachia'] == 1]
data_NA = df[df['Appalachia'] == 0]

# Filter out unexpected class labels in the target variable
valid_classes = [0, 1, 2]
data_A_filtered = data_A[data_A['Survival recode'].isin(valid_classes)]
data_NA_filtered = data_NA[data_NA['Survival recode'].isin(valid_classes)]
```

```python
# Inspect the target variable for unique values
print("Unique values in the target variable:")
print(data_A['Survival recode'].unique())
print(data_NA['Survival recode'].unique())
```

```python
# Filter out unexpected class labels in the target variable
valid_classes = [0, 1, 2]

data_A_filtered = data_A[data_A['Survival recode'].isin(valid_classes)]
data_NA_filtered = data_NA[data_NA['Survival recode'].isin(valid_classes)]

# Confirm filtering
print("Unique values after filtering - Appalachia:")
print(data_A_filtered['Survival recode'].unique())

print("Unique values after filtering - Non-Appalachia:")
print(data_NA_filtered['Survival recode'].unique())
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Features to be standardized
features = [
    'Age with single ages and 100+',
    'Year of diagnosis',
    'Sex',
    'Race',
    'Ethnicity',
    'Marital status at diagnosis',
    'Primary Site',
    'Summary Stage (2004+)'  
]

# Standardize the filtered data_A and data_NA datasets
scaler_A = StandardScaler()
X_A = scaler_A.fit_transform(data_A_filtered[features])
y_A = data_A_filtered['Survival recode'].values  # Ensure y_A is a 1D array

scaler_NA = StandardScaler()
X_NA = scaler_NA.fit_transform(data_NA_filtered[features])
y_NA = data_NA_filtered['Survival recode'].values  # Ensure y_NA is a 1D array

# Split data_A into train, test, validate
X_train_A, X_temp_A, y_train_A, y_temp_A = train_test_split(X_A, y_A, test_size=0.3, random_state=42)
X_test_A, X_validate_A, y_test_A, y_validate_A = train_test_split(X_temp_A, y_temp_A, test_size=0.33, random_state=42)

# Split data_NA into train, test, validate
X_train_NA, X_temp_NA, y_train_NA, y_temp_NA = train_test_split(X_NA, y_NA, test_size=0.3, random_state=42)
X_test_NA, X_validate_NA, y_test_NA, y_validate_NA = train_test_split(X_temp_NA, y_temp_NA, test_size=0.33, random_state=42)

print("Data filtering, standardization, and splitting complete.")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, roc_auc_score

# Solvers for logistic regression
solvers = ['newton-cg', 'liblinear', 'lbfgs']
models_A = {}
models_NA = {}

for solver in solvers:
    # Logistic Regression for data_A
    model_A = LogisticRegression(solver=solver)
    model_A.fit(X_train_A, y_train_A)
    models_A[solver] = model_A
    
    # Logistic Regression for data_NA
    model_NA = LogisticRegression(solver=solver)
    model_NA.fit(X_train_NA, y_train_NA)
    models_NA[solver] = model_NA

    # Evaluate and print metrics for both datasets
    y_pred_A = model_A.predict(X_test_A)
    y_pred_NA = model_NA.predict(X_test_NA)
    
    print(f'Logistic Regression ({solver}) - Appalachia:')
    print('Accuracy:', accuracy_score(y_test_A, y_pred_A))
    print('Confusion Matrix:\n', confusion_matrix(y_test_A, y_pred_A))
    auc_A = roc_auc_score(y_test_A, model_A.predict_proba(X_test_A)[:, 1], multi_class='ovr')
    print('ROC AUC:', auc_A)
    print()

    print(f'Logistic Regression ({solver}) - Non-Appalachia:')
    print('Accuracy:', accuracy_score(y_test_NA, y_pred_NA))
    print('Confusion Matrix:\n', confusion_matrix(y_test_NA, y_pred_NA))
    auc_NA = roc_auc_score(y_test_NA, model_NA.predict_proba(X_test_NA)[:, 1], multi_class='ovr')
    print('ROC AUC:', auc_NA)
    print()
```

```python
from xgboost import XGBClassifier

# Default XGBoost models
xgb_model_A_default = XGBClassifier()
xgb_model_A_default.fit(X_train_A, y_train_A)
xgb_model_NA_default = XGBClassifier()
xgb_model_NA_default.fit(X_train_NA, y_train_NA)

# Hyperparameterized XGBoost models
xgb_model_A_hyper = XGBClassifier(learning_rate=0.1, max_depth=5, n_estimators=100)
xgb_model_A_hyper.fit(X_train_A, y_train_A)
xgb_model_NA_hyper = XGBClassifier(learning_rate=0.1, max_depth=5, n_estimators=100)
xgb_model_NA_hyper.fit(X_train_NA, y_train_NA)

# Evaluate and print metrics for default and hyperparameterized XGBoost models
def evaluate_xgb_model(model, X_test, y_test, dataset_name, model_type):
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    cm = confusion_matrix(y_test, y_pred)
    auc = roc_auc_score(y_test, model.predict_proba(X_test)[:, 1], multi_class='ovr')
    
    print(f'XGBoost ({model_type}) - {dataset_name}:')
    print('Accuracy:', accuracy)
    print('Confusion Matrix:\n', cm)
    print('ROC AUC:', auc)
    print()

evaluate_xgb_model(xgb_model_A_default, X_test_A, y_test_A, 'Appalachia', 'Default')
evaluate_xgb_model(xgb_model_A_hyper, X_test_A, y_test_A, 'Appalachia', 'Hyperparameterized')
evaluate_xgb_model(xgb_model_NA_default, X_test_NA, y_test_NA, 'Non-Appalachia', 'Default')
evaluate_xgb_model(xgb_model_NA_hyper, X_test_NA, y_test_NA, 'Non-Appalachia', 'Hyperparameterized')
```
